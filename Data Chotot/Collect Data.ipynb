```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import time
import csv
```

```python
bedrooms_list, bathrooms_list, type_house_list, length_list, width_list, furniture_list, papers_list, price_list, address_list, direction_list = [],[],[],[],[],[],[],[],[],[]
```

```python
def parse(http):
    page = requests.get(http)
    soup = BeautifulSoup(page.text, 'html.parser')
    house_list = soup.find(class_='list-view')
    count = 1
    for i in house_list:
        for film_name in i.find_all("div", {"role": "button"}):
            if film_name.text != '':
                temp = film_name.find('a')
                URL = 'https://nha.chotot.com/'+ temp.get('href')
                new_page = requests.get(URL)
                soup = BeautifulSoup(new_page.content, 'html.parser')
                
                print('\n')
                print(count)
                count +=1
                
                width = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "width"})
                if width is not None:
                    width= width.text[:-2]
                width_list.append(width)   
                print('width: ', width)

                length = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "length"})
                if length is not None:
                    length= length.text[:-2]
                length_list.append(length)
                print('length: ', length)

                bedrooms = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "rooms"})
                if bedrooms is not None:
                    bedrooms=bedrooms.text
                bedrooms_list.append(bedrooms)
                print('bedrooms: ', bedrooms)

                bathrooms = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "toilets"})
                if bathrooms is not None:
                    bathrooms = bathrooms.text
                bathrooms_list.append(bathrooms)
                print('bathrooms: ', bathrooms)


                house_type = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "house_type"})
                if house_type is not None:
                    house_type = house_type.text
                type_house_list.append(house_type)
                print('house_type: ', house_type)

                furnishing_sell = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "furnishing_sell"})
                if furnishing_sell is not None:
                    furnishing_sell = furnishing_sell.text
                furniture_list.append(furnishing_sell)
                print('furnishing_sell: ', furnishing_sell)


                property_legal_document = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "property_legal_document"})
                if property_legal_document is not None:
                    property_legal_document= property_legal_document.text
                papers_list.append(property_legal_document)
                print('property_legal_document: ', property_legal_document)

                direction = soup.find("span", {"class": "adParamValue___25KeI", "itemprop": "direction"})
                if direction is not None:
                    direction = direction.text
                direction_list.append(direction)
                print('direction: ', direction)

                address = soup.find("span", {"class": "fz13"})
                if address is not None:
                    address = address.text
                address_list.append(address)
                print('address: ', address)


                price = soup.find("span", {"itemprop": "price"})
                if price is not None:
                    price =price.text.split('-')[0]
                price_list.append(price)
                print('price: ', price)
                
                
```

```python
text = 'https://nha.chotot.com/tp-ho-chi-minh/mua-ban-nha-dat?page='
for i in range(50):
    parse(text+str(i))
    time.sleep(0.5)
```

```python
dict_ = {'bedrooms': bedrooms_list,
        'bathrooms': bathrooms_list,
        'house_type': type_house_list,
        'length': length_list,
        'width': width_list,
        'furniture':furniture_list,
        'legal_paper': papers_list,
        'address':address_list,
        'direction': direction_list,
        'price': price_list} 

df = pd.DataFrame(dict_)
df
```

```python
df.to_csv('data.csv',quoting = csv.QUOTE_NONNUMERIC, encoding = "utf-8-sig", index= False)
```
